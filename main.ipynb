```python
# For Calculus

s = 1.5 * std_calculus + mean_calculus
a = 0.5 * std_calculus + mean_calculus
b = mean_calculus - 0.5 * std_calculus
c = mean_calculus - std_calculus
d = mean_calculus - 1.5 * std_calculus

def grade_calc(x):
    if x >= s:
        return 'S'
    elif x >= a:
        return 'A'
    elif x >= b:
        return 'B'
    elif x >= c:
        return 'C'
    elif x >= d:
        return 'D'
    else:
        return 'E'

file['Grade(Calculus)'] = file['Calculus'].apply(grade_calc)
```

```python
# For Chemistry

s = 1.5 * std_chemistry + mean_chemistry
a = 0.5 * std_chemistry + mean_chemistry
b = mean_chemistry - 0.5 * std_chemistry
c = mean_chemistry - std_chemistry
d = mean_chemistry - 1.5 * std_chemistry

def grade_chem(x):
    if x >= s:
        return 'S'
    elif x >= a:
        return 'A'
    elif x >= b:
        return 'B'
    elif x >= c:
        return 'C'
    elif x >= d:
        return 'D'
    else:
        return 'E'

file['Grade(Chemistry)'] = file['Chemistry'].apply(grade_chem)
```

```python
# Grade to point mapping
grade_points = {
    'S': 10,
    'A': 9,
    'B': 8,
    'C': 7,
    'D': 6,
    'E': 5,
    'F': 0
}

# Credits
credits = {
    'Grade(Calculus)': 4,
    'Grade(EEE)': 4,
    'Grade(Chemistry)': 4,
    'Grade(CSE)': 4,
    'Grade(English)': 2
}

# Calculate CGPA without creating any extra column
file['CGPA'] = file.apply(
    lambda row: (
        sum(grade_points[row[col]] * credits[col] for col in credits)
        / sum(credits.values())
    ),
    axis=1
)
```

```python
# Rounding off the cgpa to two decimal point

file['CGPA'] = file['CGPA'].round(2)
```
